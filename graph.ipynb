```python
x=[2,6]
y=[4,9]
plt.plot(x,y)
plt.show()
```

```python
x=[2,6]
y=[4,9]

x1=[8,3]
y1=[4,7]
plt.xlabel("Time")
plt.ylabel("speed")
plt.plot(x,y,label="road")
plt.plot(x1,y1,label="railway")
plt.legend()
plt.grid(color='y')
plt.show()
```

```python
x=[200,100,160,90]
y=["Virat Kohli","MS Dhoni","Ab De Villiers","Rohit Sharma"]
plt.xlabel("Players")
plt.ylabel("Runs")
plt.bar(y,x,label="runs")
plt.legend()
plt.grid(color='y')
plt.show()
```

```python
x=[2,6,5,6,7,78,8]
y=[4,9,23,6,9,6,5]

x1=[2,9,5,6,2,38,8]
y1=[5,8,24,5,8,2,5]

plt.xlabel("distance")
plt.ylabel("year")
plt.scatter(x,y,label="mines",marker='x',s=120,c='r')
plt.scatter(x1,y1,label="rivers",s=150)

plt.legend()
plt.grid(color='g')
plt.show()
```
